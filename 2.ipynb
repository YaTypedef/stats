```python
%matplotlib inline

from scipy.stats import expon
import math
import matplotlib.pyplot as plt

def generate(N):
    return expon.rvs(size=N)

def estimate(samples, k):
    result = []

    k_factorial = math.factorial(k)
    mean = samples[0] ** k

    for i, sample in enumerate(samples):
        mean = (i * mean + sample ** k) / (i + 1)
        error = abs(1 - (k_factorial / mean) ** (1 / float(k)))
        result.append(error)
    return result

def draw(N, k_values, max_y):
    samples = generate(N)

    plt.figure(figsize=(15,10))
    plt.axis([0, N, 0] + [max_y])

    for k in k_values:
        plt.plot(range(N), estimate(samples, k), label=str(k))

    plt.legend()
    plt.show()
```

Сначала построим графики для небольших значений k:

```python
draw(10 ** 4, [1, 3, 5, 6, 8], 0.1)
```

Теперь увеличим масштаб по оси ординат и построим графики для бОльших значний k:

```python
draw(10 ** 4, [10, 20, 50, 70, 100], 4)
```
