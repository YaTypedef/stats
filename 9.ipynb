```python
# encoding: utf-8
%matplotlib inline

from scipy.stats import bernoulli
import numpy as np
import matplotlib.pyplot as plt

SAMPLE_SIZE = 1000
SAMPLE_COUNT = 500

def calc_vars(thetas):
    vars = []
    for theta in thetas:
        samples = bernoulli.rvs(theta, size=SAMPLE_SIZE)
        mean = np.mean(samples)
        vars.append(np.var([np.mean(bernoulli.rvs(mean, size=SAMPLE_SIZE))
                            for _ in range(SAMPLE_COUNT)]))
    return vars



plt.figure(figsize=(15,10))

thetas = np.arange(0, 1.01, 0.01)
plt.plot(thetas, thetas * (1 - thetas) / SAMPLE_SIZE)
plt.plot(thetas, calc_vars(thetas))

plt.show()
```

Заметим, что бутстрепная оценка дисперсии совпадает с нижней оценкой дисперсии из неравенства Рао-Крамера, поскольку мы взяли эффективную оценку.
